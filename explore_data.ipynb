```python
import glob
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random
# Visualize some images in the dataset

# Large dataset
img_cars_files = glob.glob("dataset\\vehicles\\**\\*.png",recursive= True)
img_not_cars_files = glob.glob("dataset\\non-vehicles\\**\\*.png",recursive= True)

img_cars = []
img_not_cars = []
for image_file in img_cars_files:
    img_cars.append(image_file)
for image_file in img_not_cars_files:
    img_not_cars.append(image_file)

grid_size = [10, 22]
grid_n_elements = grid_size[0]*grid_size[1]
fig, axs = plt.subplots(grid_size[0], grid_size[1], figsize=(12,6))
fig.subplots_adjust(hspace = .001, wspace=.001)
axs = axs.ravel()

random.shuffle(img_cars)
random.shuffle(img_not_cars)
print("Example of car images")
for i, im in enumerate(img_cars[0:grid_n_elements]):
    axs[i].imshow(mpimg.imread(im))
    axs[i].axis('off')
plt.savefig("cars.png")
plt.show()
print("Example of non-car images")
fig, axs = plt.subplots(grid_size[0], grid_size[1], figsize=(24,12))
fig.subplots_adjust(hspace = .001, wspace=.001)
axs = axs.ravel()
for i, im in enumerate(img_not_cars[0:grid_n_elements]):
    axs[i].imshow(mpimg.imread(im))
    axs[i].axis('off')
plt.savefig("non-cars.png")
plt.show()
```

```python
# Show YUV color space on a test image
import cv2
import numpy as np

im_file = "test_images\\test4.jpg"
img = mpimg.imread(im_file)
imgcpy = np.copy(img)
grid_size = [1, 3]
fig, axs = plt.subplots(grid_size[0], grid_size[1], figsize=(24,8))
fig.subplots_adjust(hspace = .001, wspace=.001)

axs[0].imshow(imgcpy)
axs[0].axis('off')
axs[1].imshow(cv2.cvtColor(imgcpy, cv2.COLOR_RGB2YUV))
axs[1].axis('off')
axs[2].imshow(cv2.cvtColor(imgcpy, cv2.COLOR_RGB2HLS)[:,:,2])
axs[2].axis('off')
plt.savefig("Color spaces.png")
plt.show()
```
